```python
import pandas as pd
import random
import numpy as np
```

```python

def generate_dummy_customer_data(num_records):
    age_min, age_max = 18, 60
    income_mean = 50000 
    income_std = 25000 
    genders = ['Male', 'Female']
    gender_weights = [0.5, 0.5]
    membership_min, membership_max = 1, 9
    locations = ['Urban', 'Rural', 'Suburban']
    location_weights = [0.4, 0.35, 0.25] 
    churn_rate = 0.10
    
    data = {}
    
    data['Age'] = [random.randint(age_min, age_max) for _ in range(num_records)]
    
    incomes = np.random.normal(loc=income_mean, scale=income_std, size=num_records)
    data['AnnualIncome'] = [round(max(10000, income), 2) for income in incomes]

    data['Gender'] = random.choices(genders, weights=gender_weights, k=num_records)

    data['MembershipDuration'] = [random.randint(membership_min, membership_max) for _ in range(num_records)]

    data['Location'] = random.choices(locations, weights=location_weights, k=num_records)


    df = pd.DataFrame(data)
    return df
```

```python
def generate_drift_customer_data(num_records):
    membership_min, membership_max = 1, 9
    genders = ['Male', 'Female']
    locations = ['Urban', 'Rural', 'Suburban']
    location_weights = [0.4, 0.35, 0.25] 

    age_min, age_max = 20, 70
    
    income_mean_drift = 65000 
    income_std = 25000 
    
    churn_rate_drift = 0.18
    
    data = {}
    
    data['Age'] = [random.randint(age_min, age_max) for _ in range(num_records)]
    
    incomes = np.random.normal(loc=income_mean_drift, scale=income_std, size=num_records)
    data['AnnualIncome'] = [round(max(15000, income), 2) for income in incomes]

    data['Gender'] = random.choices(genders, k=num_records)

    data['MembershipDuration'] = [random.randint(membership_min, membership_max) for _ in range(num_records)]

    data['Location'] = random.choices(locations, weights=location_weights, k=num_records)


    df = pd.DataFrame(data)
    return df
```

```python
df = generate_dummy_customer_data(1000)
df.to_csv("../artifact/dummy_inference_churn_data.csv",index=False)
```

```python
df = generate_drift_customer_data(1000)
df.to_csv("../artifact/dummy_inference_churn_drift_data.csv",index=False)
```
